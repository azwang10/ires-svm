```python
import pandas as pd

positive_frame = pd.read_csv('./data/data_positive_hash.csv', index_col=0)
negative_frame = pd.read_csv('./data/data_negative_hash.csv', index_col=0)
```

```python
import numpy as np

positive_at = np.array([(i.count('A') + i.count('T')) / len(i) for i in positive_frame.index])
negative_at = np.array([(i.count('A') + i.count('T')) / len(i) for i in negative_frame.index])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('poster')
import numpy as np

sns.violinplot(data=[positive_at, negative_at], scale='count')
plt.xticks(np.arange(2), ('Positive', 'Negative'))
plt.show()
```
